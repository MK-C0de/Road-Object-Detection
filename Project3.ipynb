```python
import cv2
import numpy as np
import copy
import random

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
print(cv2.__version__)
```

```python
x1_path = '/content/drive/MyDrive/EEL 6990 Computer Vision/EEL 6990 Project 3/content/yolov5x1-night-stop.mp4'
x20_path = '/content/drive/MyDrive/EEL 6990 Computer Vision/EEL 6990 Project 3/content/yolov5x20-night-stop.mp4'

x1_out = '/content/drive/MyDrive/EEL 6990 Computer Vision/EEL 6990 Project 3/night-stop-1-3.jpg'
x20_out = '/content/drive/MyDrive/EEL 6990 Computer Vision/EEL 6990 Project 3/night-stop-20-3.jpg'

video_1 = cv2.VideoCapture(x1_path)
video_20 = cv2.VideoCapture(x20_path)

count = 0

sample1 = []
sample20= []
sample_frame= []

while True:
  ret1, frame1 = video_1.read()
  ret20, frame20 = video_20.read()

  if frame1 is None or frame20 is None: break

  if ret1 and ret20:
    if random.random() < 0.038:
        sample1.append(frame1)  
        sample20.append(frame20)  
        
        sample_frame.append(count) 
    count += 1

video_1.release()
video_20.release()

print('Number of Sample Frames: ', len(sample_frame))
```

```python
fig = plt.figure(figsize=(25, 35))
rows, columns = 15, 7

sample_std = []
for frame in sample1:
  height, width = frame.shape[0], frame.shape[1]
  rgb = cv2.cvtColor(np.copy(frame), cv2.COLOR_BGR2RGB)

  sample_std.append(rgb)

frame_check = 0
for count in range(len(sample_std)):
  fig.add_subplot(rows, columns, count+1)
  plt.imshow(sample_std[count])
  plt.axis('off')
  plt.title(f'{sample_frame[frame_check]}')
  frame_check += 1

plt.savefig(x1_out, bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(25, 35))
rows, columns = 15, 7

sample_std2 = []
for frame in sample20:
  height, width = frame.shape[0], frame.shape[1]
  rgb = cv2.cvtColor(np.copy(frame), cv2.COLOR_BGR2RGB)

  sample_std2.append(rgb)

frame_check = 0
for count in range(len(sample_std2)):
  fig.add_subplot(rows, columns, count+1)
  plt.imshow(sample_std2[count])
  plt.axis('off')
  plt.title(f'{sample_frame[frame_check]}')
  frame_check += 1

plt.savefig(x20_out, bbox_inches='tight')
plt.show()
```
